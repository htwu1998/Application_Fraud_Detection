```python
from datetime import datetime
start_time = datetime.now()

# Load libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

print('LOAD DURATION:',datetime.now() - start_time)
```

```python
%%time
data = pd.read_csv('applications data.csv', parse_dates = ['date','dob'])
```

```python
# getting more stats
cat_name = ["ssn", "record", "zip5", "homephone"]
for col in cat_name:
    data[col] = data[col].astype('object')
data['fraud_label'] = data['fraud_label'].astype('category')

statistics_of_data = []
for col in data.columns:
      statistics_of_data.append((col,
                             data[col].count(),
                             (data[col].notnull().sum() - data[col].isnull().sum())*100/data.shape[0],
                             data[col].nunique(),
                             data[col].dtype,
                             data[col].mode()[0]
                             ))
stats_df = pd.DataFrame(statistics_of_data, columns=['Feature', '# of records','% populated','Unique Values', 'Type','Most Common Field Value'])
stats_df
```

```python
# visualizing each field
def population_ratio(df, col, n):
    fig, ax = plt.subplots(figsize=(20,10))
    X = (df[col].value_counts()[:n]*100/data.shape[0])
    X.plot(kind='bar')
    plt.title(f'{col} field population observation', fontsize=15)
    ax.set_xlabel(f'{col}', fontsize=15)
    ax.set_ylabel(f'Percentage of Population', fontsize=15)
    plt.show()
    fig.savefig(f'{col}_ratio.png')

    
def count_plot(df,col,n):
    fig, ax = plt.subplots(figsize=(20,10))
    X = df[col].value_counts()[:n]
    X.plot(kind='bar')
    plt.title(f'{col} field values count', fontsize=15)
    ax.set_xlabel(f'{col}', fontsize=15)
    ax.set_ylabel(f'Count Number', fontsize=15)
    plt.show()
    fig.savefig(f'{col}_count.png')
    
def log_scaled_plot(df,col,n):
    fig, ax = plt.subplots(figsize=(20,10))
    X = df[col].value_counts()[:n]
    X.plot(kind='bar', logy=True)
    plt.title(f'{col} log scaled field values count', fontsize=15)
    ax.set_xlabel(f'{col}', fontsize=15)
    ax.set_ylabel(f'Log Scaled Count', fontsize=15)
    plt.show()
    fig.savefig(f'{col}_log.png')
```

```python
# date
count_plot(data,'date',20)
```
